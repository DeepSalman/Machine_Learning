```python
def linear_regression(x,y):
    if len(x)!=len(y) or len(x)==0 or len(y)==0:
        return -1
    
    n=len(x)

    mean_x=sum(x)/n
    mean_y=sum(y)/n

    #Calculate slope

    num = sum((x[i]-mean_x)*(y[i]-mean_y) for i in range(n))
    den = sum((x[i]-mean_x)**2 for i in range(n))

    slope=num/den
    intercept=mean_y-slope*mean_x

    return slope,intercept 
```
