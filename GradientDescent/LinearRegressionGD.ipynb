```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sb
```

```python
import numpy as np

class LinearRegressionGD:
    def __init__(self,alpha,epoch):
        self.alpha = alpha
        self.epoch = epoch
        self.theta = None
        self.cost_history = []

    def fit(self, X, y):
        X, y = np.array(X), np.array(y)
        m, n = X.shape
        self.theta = np.zeros(n)

        for _ in range(self.epoch):
            predictions = X @ self.theta
            errors = predictions - y
            gradient = (1/m) * (X.T @ errors)
            self.theta -= self.alpha * gradient
            cost = (1/(2*m)) * np.sum(errors**2)
            self.cost_history.append(cost)

        self.theta = np.round(self.theta, 4)
        return self

    def predict(self, X):
        X = np.array(X)
        if self.theta is None:
            raise Exception("Model not trained yet.")
        if X.shape[1] != len(self.theta):
            raise ValueError("Feature size mismatch.")
        return X @ self.theta

    def get_coefficients(self):
        return self.theta
```

```python
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=200,n_features=2,n_informative=2,n_targets=1,noise=50)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model = LinearRegressionGD(0.01,200)
```

```python
model.fit(x_train,y_train)
```

```python
print("Coef ",model.get_coefficients())
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
print(score)
```
